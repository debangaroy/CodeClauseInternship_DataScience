# Exploratory Data Analysis (EDA) on Iris Dataset

### Task Overview
The objective of this task is to perform exploratory data analysis (EDA) on the Iris dataset, which includes loading the dataset, obtaining basic statistics, checking for missing values, and visualizing the data through various plots. This analysis helps in understanding the distribution and relationships of different features in the dataset.

#### Importing Libraries

```python
#Importing Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

#### Loading the Iris Dataset

```python
#Loading the Iris Dataset
iris_df=sns.load_dataset("iris")
iris_df.head()
```

##### Data Summary

```python
# Get a summary of the dataframe
iris_df.info()
```

```python
# Display basic statistics of the dataset
print(iris_df.describe())
```

```python
# Check for missing values
iris_df.isnull().sum()
```

##### Distribution of Each Species Name

```python
# Distribution of each species name
print(iris_df['species'].value_counts())

# Plot the Distribution of each species
plt.figure(figsize=(8,6))
sns.countplot(x='species', data=iris_df)
plt.show()
```

#### Line Plot of Sepal Length vs Sepal Width

```python
plt.figure(figsize=(8,6))
sns.lineplot(x='sepal_length',y='sepal_width',data=iris_df,hue='species')
plt.show()
```

*Insights*:

The plot helps to visually identify differences in the sepal dimensions among the three species.

It shows Setosa can be distinguished from Versicolor and Virginica based on sepal width for given sepal lengths.

Versicolor and Virginica show more similarity and overlap in their sepal dimensions, making them harder to distinguish based solely on these two features.

#### Scatter Plot of Petal Length vs Petal Width

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='petal_length', y='petal_width',data=iris_df, hue='species')
plt.show()
```

*insights*:

This scatter plot clearly shows how the three species of Iris can be distinguished based on their petal length and width.

Setosa is easily distinguishable due to its much smaller petal dimensions.

Versicolor and Virginica can be distinguished based on their petal dimensions, Virginica having larger petal dimensions than Versicolor.

#### Histograms of Features

```python
# Histogram for sepal length
sns.displot(iris_df['sepal_length'],kde=False)
plt.title('Histogram of Sepal Length')
plt.show()

# Histogram for sepal width
sns.displot(iris_df['sepal_width'],kde=False)
plt.title('Histogram of Sepal Width')
plt.show()

# Histogram for petal length
sns.displot(iris_df['petal_length'],kde=False)
plt.title('Histogram of Peat Length')
plt.show()

# Histogram for petal width
sns.displot(iris_df['petal_width'],kde=False)
plt.title('Histogram of Petal Width')
plt.show()
```

Sepal Length: 
Most common length is around 5 to 6 cm.

Sepal Width: 
Two common widths, around 2.5 to 3.5 cm.

Petal Length: 
Most flowers have short petals (2 to 3 cm), but there is a wide range up to 7 cm.

Petal Width: 
Most flowers have narrow petals (0.1 to 0.6 cm), but widths can go up to 2.5 cm

#### KDE Plots of Features

```python
# KDE plot for sepal length
sns.displot(iris_df['sepal_length'], kind="kde", fill=True)
plt.title('Distribution of Sepal Length')
plt.show()

# KDE plot for sepal width
sns.displot(iris_df['sepal_width'], kind="kde", fill=True)
plt.title('Distribution of Sepal Width')
plt.show()

# KDE plot for petal length
sns.displot(iris_df['petal_length'], kind="kde", fill=True)
plt.title('Distribution of Petal Length')
plt.show()

# KDE plot for petal width
sns.displot(iris_df['petal_width'], kind="kde", fill=True)
plt.title('Distribution of Petal Width')
plt.show()
```

Sepal Length: 
The distribution of sepal length appears normal (bell-shaped) with a height of about 5.5 to 6.0 units.
It shows a slight skewness towards longer sepal lengths.

Sepal Width:
The distribution of sepal width is almost normal but wide compared to sepal length.
It peaks around 3.0 units and shows a high frequency trend between 2.8 and 3.4 units.

Petal Length:
Petal length display bimodal distribution with two peaks: one around 1.5 units and another around 4.5 to 6.0 units.
This indicates two distinct groups of iris flowers based on their petal lengths.

Petal Width:
Petal width also shows a bimodal distribution with peaks around 0.25 units and 1.5 units.
Similar to petal length, this suggests two different groups of iris flowers based on their petal widths.

#### Boxplots of Features

```python
# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris_df)
plt.show()

# Boxplot
sns.boxplot(data=iris_df, x='species', y='sepal_length')
plt.show()

# Boxplot
sns.boxplot(data=iris_df, x='species', y='sepal_width')
plt.show()

# Boxplot
sns.boxplot(data=iris_df, x='species', y='petal_length')
plt.show()

# Boxplot
sns.boxplot(data=iris_df, x='species', y='petal_width')
plt.show()
```

**Overall Boxplot of All Features**:

Sepal Length: Most flowers have sepal lengths between 5.1 and 6.5, with a few very short or very long ones.

Sepal Width: Most flowers have sepal widths between 2.8 and 3.4, with a few very narrow or very wide ones, and some outliers.

Petal Length: Petal lengths vary widely, with most between 1.7 and 5.1. There are significant differences among flowers.

Petal Width: Petal widths also vary a lot, mostly between 0.5 and 1.7. There are a few extremely narrow or wide ones.

**Sepal Length by Species**:

Setosa species has the shortest sepal lengths, with minimal variation.

Versicolor and virginica species have longer sepal lengths, with virginica having the highest median.

**Sepal Width by Species**:

Setosa species has the widest sepals on average.

Versicolor and virginica species have similar sepal widths, with slightly more variation in virginica.

**Petal Length by Species**:

Setosa species has the shortest petal lengths with very little variation.

Versicolor and virginica show a clear distinction, with virginica having the longest petals.

**Petal Width by Species**:

Setosa species has the narrowest petal widths.

There is a clear separation between versicolor and virginica, with virginica having the widest petals.


#### Additional Scatter Plots

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris_df,hue='petal_length')
plt.show()
sns.scatterplot(x='sepal_width',y='petal_width',data=iris_df,hue='petal_width')
plt.show()

sns.scatterplot(x='sepal_length',y='petal_length',data=iris_df,hue='sepal_length')
plt.show()
sns.scatterplot(x='sepal_width',y='petal_width',data=iris_df,hue='sepal_width')
plt.show()
```

**Overall Analysis**:

**Strong Correlation**: Sepal length and petal length are strongly correlated. Flowers with longer sepals generally have longer petals.

**Moderate Correlation**: Sepal width and petal width show a moderate positive correlation. The relationship is not as strong as that of sepal and petal lengths.

**Variation**: There is more variation in petal dimensions compared to sepal dimensions.

#### Mean Values of Features

```python
print(iris_df['sepal_length'].mean())
print(iris_df['sepal_width'].mean())
print(iris_df['petal_length'].mean())
print(iris_df['petal_width'].mean())
```

```python
# Calculate the mean of each feature
feature_means = iris_df.drop(['species'], axis=1).mean()

# Plot the mean values as a bar chart
feature_means.plot(kind='bar', figsize=(10, 6))
plt.ylabel('Mean Value')
plt.title('Mean of Each Feature in the Iris Dataset')
plt.show()
```

**Insights from Bar Chart**:

Sepals are generally longer and wider than petals.

Petal lengths are significantly longer than petal widths.

Among all features, sepal length is the longest on average, and petal width is the shortest.

#### Distribution of Species as a Pie Chart

```python
# Distribution of species as a pie chart
species_counts = iris_df['species'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Species in the Iris Dataset')
plt.show()
```

**Insights from Pie Chart**:

The Iris dataset is evenly balanced among its three species: Setosa, Versicolor, and Virginica.

Each species accounts for exactly one-third of the total dataset, demonstrating a perfect balance in species representation.

#### Joint Plots

```python
sns.jointplot(x='sepal_length',y='petal_length',data=iris_df,kind='reg')
plt.show()
sns.jointplot(x='sepal_width',y='petal_width',data=iris_df,kind='reg')
plt.show()

sns.jointplot(data=iris_df, x='sepal_length', y='sepal_width', hue='species')
plt.show()
sns.jointplot(data=iris_df, x='petal_length', y='petal_width', hue='species')
plt.show()
```

*Sepal Length vs. Petal Length (with Regression Line)*:
Flowers with longer sepals tend to have longer petals.

*Sepal Width vs. Petal Width (with Regression Line)*:
Flowers with wider sepals generally have wider petals, but the relationship is not as strong as that between sepal length and petal length.

*Sepal Length vs. Sepal Width (colored by Species)*:
Sepal length and width vary independently, and each species has distinct sepal dimensions.

*Petal Length vs. Petal Width (colored by Species)*:
Petal length and width are strongly related, and each species can be clearly differentiated based on petal dimensions.

#### Pair Plot

```python
sns.pairplot(iris_df,hue='species',markers=["o", "s", "D"])
plt.show()
```
